```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.sosracismomadrid.es/'
sos = requests.get(url).text
print(sos)
```

```python
soup = bs(sos, "html.parser")
soup
```

```python
headers = soup.find_all("h2")
print(headers)
```

```python
list_ = []

for h2 in headers:
    list_.append(h2.text)

print(list_)
```

```python
import pandas as pd

df = pd.DataFrame({'Headers': list_})
print(df)
```

```python
imgs = soup.find_all("img")
print(imgs)
```

```python
images = []
for img in imgs:
    src = img.get("src")
    images.append(src)

print(images)
```

```python
df_i = pd.DataFrame({'Image Urls': images})
print(df_i)
```

```python
import sqlite3 as sq
con = sq.connect('sos.db')
df_i.to_sql('images',con)
con.commit()
```

```python
con.close()
con = sq.connect('sos.db')
df.to_sql('titulos', con)
```
